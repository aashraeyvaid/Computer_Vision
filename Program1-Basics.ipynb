```python
import cv2
print(cv2.__version__)
```

```python
image=cv2.imread('./images/flowerpot1.jpg')
```

```python
import matplotlib.pyplot as plt

plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
def show(title="",image=None):
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
show('Flower Pot',image)
```
