```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show(title="",image=None,size=10):
    w,h=image.shape[0],image.shape[1]
    aspect_ratio=w/h
    plt.figure(figsize=(size*aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
image=cv2.imread('./images/flowerpot2.jpg')
```

```python

b,g,r=cv2.split(image)
show(image=b)
```

```python
zeros=np.zeros(image.shape[:2],dtype='uint8')

show('red',cv2.merge([zeros,zeros,r]))
show('green',cv2.merge([zeros,g,zeros]))
show('blue',cv2.merge([b,zeros,zeros]))
```

```python
merged=cv2.merge([r,g,b])
show('',merged)
```

```python
show('',cv2.merge([b+100,g,r]))
```

```python
show('',cv2.merge([b,g,r]))
```

```python
hsv_image=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
show('',hsv_image)
```
